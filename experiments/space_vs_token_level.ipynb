```python
data_dir = "/home/aziz/experiments/output/gans_for_apr/"


with open(data_dir+"space_level.txt", 'r', encoding='utf-8') as f:
    space_text = f.read()
with open(data_dir+"token_level.txt", 'r', encoding='utf-8') as f:
    token_text = f.read()

space_lines = space_text.split('\n')
token_lines = token_text.split('\n')

gtsw = "Fixed code: "
ground_truth = [x.replace(gtsw, "", 1) for x in space_lines if x.startswith(gtsw)]
gen_sw = "Genration:  "
space_generation = [x.replace(gen_sw, "", 1) for x in space_lines if x.startswith(gen_sw)]
token_generation = [x.replace(gen_sw, "", 1) for x in token_lines if x.startswith(gen_sw)]
```

```python
for gt, sl, tl in zip(ground_truth, space_generation, token_generation):
    print("Ground Truth: ", gt)
    print("--")
    print("Token-level:  ", tl)
    print("--")
    print("Space-level:  ", sl)
    print("==============")
    print("==============")
```
